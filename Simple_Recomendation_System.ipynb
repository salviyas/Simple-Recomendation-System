```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data={
    'User':['A','A','A','B','B','C','C','D','D','D'],
    "Item":['Item1','Item2',"Item3","Item1","Item2","Item2","Item3","Item1","Item2","Item3"],
    "Rating":[5,3,4,4,2,5,3,2,4,3]
}

df=pd.DataFrame(data)
df
```

```python
user_item_matrix=df.pivot_table(index='User',columns='Item',values='Rating')
user_item_matrix
```

```python
user_similarity=cosine_similarity(user_item_matrix.fillna(0))# fill the missing values weith 0
user_similarity_df=pd.DataFrame(user_similarity,index=user_item_matrix.index,columns=user_item_matrix.index)

user_similarity_df
```

```python
def get_recommendations(user,user_item_matrix,user_similarity_df):
  similar_users=user_similarity_df[user]

  similar_user_ratings=user_item_matrix.T.dot(similar_users)

  similarity_sum=similar_users.sum()

  recomended_items=similar_user_ratings/similarity_sum

  recomended_items=recomended_items[user_item_matrix.loc[user].isna()]

  return recomended_items.sort_values(ascending=False)



recommendations=get_recommendations('C',user_item_matrix,user_similarity_df)
recommendations
```
